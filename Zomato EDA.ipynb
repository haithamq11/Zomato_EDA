```python
# Import libraries 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import matplotlib
```

```python
# Read zomato dataset
df = pd.read_csv('zomato.csv', encoding='latin-1')
df.head()
```

```python
# Details of df
df.info()
```

```python
df.describe()
```

```python
# Check the null counts
df.isnull().sum()
```

```python
# Visualize the null values
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.heatmap(df.isnull(),yticklabels=False, cbar=False, cmap='viridis')
```

```python
# Read the part 2 of zomato
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# combine the two dataset by'Country Code'
final_df = pd.merge(df,df_country,on='Country Code', how='left')
final_df.head()
```

```python
# show the countries
country_name = final_df['Country'].value_counts().index
country_name
```

```python
# Display the branches of zomato
country_values = final_df['Country'].value_counts().values
country_values
```

```python
plt.pie(country_values[:3],labels=country_name[:3],autopct='%1.2f%%', shadow=True, explode=[0.2,0.2,0.2])
```

```python
rating = final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:'Rate counts'})
rating
```

```python
sns.barplot(x='Aggregate rating', y='Rate counts', data=rating, hue= 'Rating color', palette=['blue','red','orange','yellow','green','green'])
```

```python
sns.countplot(x='Rating color', data=rating, palette=['blue','red','orange','yellow','green','green'])
```
